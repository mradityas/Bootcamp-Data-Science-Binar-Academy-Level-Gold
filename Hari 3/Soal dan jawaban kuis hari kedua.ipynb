```python
#2
df_male = insurance.loc[insurance['sex'] == 'male']
df_group_male = df_male.groupby('sex').agg({'bmi' : ['min','max'], 'charges' : 'mean'})
df_group_male
```

```python
#3
def a (b):
    if b < 18.5 :
        return ('Berat Badan Kurang')
    elif b > 19.5 and b < 22.9 :
        return ('Berat Badan Normal')
    elif b > 23 and b < 24.9 :
        return ('Berat Badan Kelebihan')
    elif b > 25 and b < 29.9 :
        return ('Obesitas')
    else : 
        return ('Obesitas II')

insurance ['kategori'] = insurance['bmi'].apply(lambda x : a(x))
insurance.head()
```

```python
obesitas2 = insurance.loc[insurance['kategori']== 'Obesitas II']
max_obesitas_2 = obesitas2.groupby('kategori').agg({'charges' :['max','min']})
max_obesitas_2
```
